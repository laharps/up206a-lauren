It is more useful for me to print out the column names to determine what variables I want to further explore.

```python
hps_cols = list(hps.columns)
hps_cols
```

The column names have been stored in the variable `hps_cols`.<br>
In order to dive further into the dataset, I will trim the data. The first 15 variables are demographic data for each respondent, which will be useful to explore.

```python
# create list of first 15 columns
cols15 = hps_cols[0:14]
# subset hps to first 15 columns
hps_trim = hps[cols15].copy()
hps_trim
```

Now that the data is trimmed to the first 15 columns, I can look further into some variables.

I'll make a histogram of the birth year of respondants. 

```python
hps_trim.TBIRTH_YEAR.plot.hist(bins = 25, title = "Frequency of Respondent Birth Year",
                              x = "Birth Year",
                              y = "Frequency")
```

This histogram shows a nearly normal distribution of ages in the respondents. <br>
<br>
Now let's look at the frequency of respondent's race using the `RRACE` variable.

```python
race = hps.RRACE.value_counts()
race
```

In order to make sense of this, I can use the data dictionary to print out the number and percentage of respondents for each race.


```python
# get number of observations to calculate percentage
l = len(hps)
# print string with percentage calculation
print("White: " + str(race[1]) + " respondents (" + str(round((race[1]/l)*100, 1)) + "%)")
print("Black: " + str(race[2]) + " respondents (" + str(round((race[2]/l)*100, 1)) + "%)")
print("Asian: " + str(race[3]) + " respondents (" + str(round((race[3]/l)*100, 1)) + "%)")
print("Other: " + str(race[4]) + " respondents (" + str(round((race[4]/l)*100, 1)) + "%)")
```

The survey respondents are disproportionately white. There is an additional variable that indicates whether the respondent is Hispanic, which would need to be incorporated for a fuller picture.  

Finally, I'll query the dataset to return only respondents from California. California's FIPS code is 06, though as we saw from using the `.head`, this dataset does not include a leading zero for the FIPS code. 

```python
hps_ca = hps[hps["EST_ST"] == 6]
```

And to confirm that `hps_ca` only contains observations from California, I'll use `.head` again.

```python
hps_ca.head()
```
